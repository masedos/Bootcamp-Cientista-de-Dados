Para esse desafio temos dois tipos de questões diferentes. Algumas sobre o Stack Overflow Annual Survey para serem resolvidas com o Power BI, e outras para serem resolvidas com consultas SQL.
1. Para as questões de relativas aos dados do Stack Overflow Survey, instale o [Power BI Desktop](https://powerbi.microsoft.com/pt-br/downloads/), baixe o dataset [Stack Overflow Annual Developer Survey - 2020](https://insights.stackoverflow.com/survey) e carregue os dados no Power BI. É um dataset com diversas colunas, para nossa atividade usaremos as seguintes: Age, Age1stCode, ConvertedComp, Country, EdLevel, JobSat, OpSys e YearsCode.

Para evitar a necessidade de instalação um servidor de banco de dados, acesse o site [SQL Online IDE](https://sqliteonline.com/), vamos utilizar o simulador do PostgreSQL.
